```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

df = pd.read_csv("../data/Hospital ER_Data.csv")
```

```python
# View Dataset

df.head()
```

```python
# Check Dataset Info

df.info()
```

```python
# Remove Unnecessary Columns

df = df.drop(["Patient First Inital", "Patient Last Name"],axis=1)
```

```python
# Fill missing department values
df["Department Referral"] = df["Department Referral"].fillna("Unknown")

# Fill missing satisfaction scores with average
df["Patient Satisfaction Score"] = df["Patient Satisfaction Score"].fillna(
    df["Patient Satisfaction Score"].mean()
)

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# cleaned data

df.to_csv("../data/hospital_er_cleaned.csv", index=False)
```

```python
# Convert Date Column

df["Patient Admission Date"] = pd.to_datetime(df["Patient Admission Date"])
```

```python
# Average Waittime

df["Patient Waittime"].mean()
```

```python
# Department Workload

df["Department Referral"].value_counts()
```

```python
# Visualization

plt.figure(figsize=(10,5))
sns.countplot(x="Department Referral", data=df)
plt.title("Patients per Department")
plt.xticks(rotation = 45)
plt.show()
```

```python
# Waittime Distribution

sns.histplot(df["Patient Waittime"],bins=20)
plt.title("Patient Wait Time Distribution")
plt.show()
```

```python
# Satisfaction vs Waittime

sns.scatterplot(
    x ="Patient Waittime",
    y ="Patient Satisfaction Score",
    data = df
)

plt.title("Wait Time vs Satisfaction")
plt.show()
```

```python
# Pateint Age Distribution

plt.figure(figsize=(8,5))

sns.histplot(df["Patient Age"], bins=20)

plt.title("Patient Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Patients")

plt.show()
```

```python
# Admission rate

sns.countplot(x="Patient Admission Flag", data=df)
plt.title("Patient Admission vs Discharge")
plt.show()
```

```python
# Peak hospital hours analysis

df["Hour"] = df["Patient Admission Date"].dt.hour
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x="Hour", data=df)

plt.title("Patient Arrival by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Patients")

plt.show()
```

```python
# Age group segmentation

bins = [0,18,35,50,65,100]

labels = [
    "Child",
    "Young Adult",
    "Adult",
    "Middle Age",
    "Senior"
]

df["Age Group"] = pd.cut(
    df["Patient Age"],
    bins=bins,
    labels=labels
)
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Age Group",
    data=df,
    order=df["Age Group"].value_counts().index
)

plt.title("Patients by Age Group")

plt.show()
```
